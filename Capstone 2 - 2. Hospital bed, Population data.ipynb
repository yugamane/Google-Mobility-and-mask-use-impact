### 1G. Web scraping Hospital bed table using beautiful soup

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.aha.org/statistics/fast-facts-us-hospitals")
soup = BeautifulSoup(page.content, 'html.parser')
soup.get_text
html_table=soup.find('tbody')
all_rows = html_table.find_all('tr')
```

```python
for pos,rows in enumerate(all_rows):
    print(pos)
    print(rows)
```

```python
 final_data_list =[]
```

```python
for pos,rows in enumerate(all_rows):
    
    col_list = [each_col.get_text(strip=True) for each_col in rows.find_all('td')]
    final_data_list.append(col_list)
```

```python
print( final_data_list)
```

```python
hosp = pd.DataFrame( final_data_list).dropna().rename(columns={0:'Description', 1: 'Count'})
hosp.head()
```
